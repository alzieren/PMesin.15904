# Operator

### Operator Aritmetika

| Operator | Arti                                                                                          | Contoh                  |
|----------|--------------------------------------------------------------------------------------------------|--------------------------|
| +        | Penambahan                                                                   | x + y                 |
| -        | Pengurangan                                              | x - y                  |
| *        | Perkalian                                                                            | x * y                    |
| /        | Pembagian                                 | x / y                    |
| %        | Modulo - Sisa Pembagian                                 | x % y (sisa pembagian of x/y) |
| //       | Pembagian dan dibulatkan ke bawah | x // y                   |
| **       | Eksponen - Pangkat                                             | x**y (x pangkat y)  |

````python
print(2 + 3)    # penjumlahan
print(2 - 3.)   # pengurangan
print(2. * 3.)  # perkalian
print(2. / 3)   # pembagian
print(2 % 3)    # modulo - sisa pembagian
````

````python
print(20 / 6)
print(20 // 6)

print(12 / 6.)
print(12 // 6.)

print(-8 + 8)
print(-8. + 16)
````

### Urutan Operator

| Prioritas| Operator |        |
|----------|----------|--------|
| 1        | +, -     | unary  |
| 2        | **       |        |
| 3        | *,/,//,% |        |
| 4        | +, -     | binary |

````python
8 + 12 *20
````

````python
print(18 % 12 % 4)
print(18 % 12)
print(6 % 4)

print(1 ** 2 ** 3)
print(2 ** 3)
print(2 ** 3)
````

````python
#variabel dan jumlah apel
Darjo = 2
Darno = 5
Waginem = 6

#cetak nama variabel dan jumlah apel
print("Darjo,", Darjo)
print("Darno,", Darno)
print("Waginem,", Waginem)
#variabel totalApel (penjumlahan semua apel)
totalApel = Darjo + Darno + Waginem
print("totalApel,", totalApel)

#otak-atik kode dengan variabel baru
apelDimakan = 4
#sisa apel
sisaDarjo = Darjo - 1
sisaDarno = Darno - 2
sisaWaginem = Waginem - 1

print("\nSetelah beberapa apel dimakan:")
print("Darjo,", sisaDarjo)
print("Darno,", sisaDarno)
print("Waginem,", sisaWaginem)

# total apel setelah dimakan
totalSisa = sisaDarjo + sisaDarno + sisaWaginem
print("totalSisa,", totalSisa)

#operator aritmatik lainnya
sisaDarjo *= 2   # kali 2
sisaDarno += 2   # tambah 2
sisaWaginem -= 1 # kurang 1

print("\nSetelah ditambah/dikurangi:")
print("Darjo,", sisaDarjo)
print("Darno,", sisaDarno)
print("Waginem,", sisaWaginem)

totalBaru = sisaDarjo + sisaDarno + sisaWaginem
print("totalBaru,", totalBaru)
````

## Latihan : Simple input and output
Coba utak atik perintah kode dibawah ini
``` python
# Masukkan nilai a dalam float
# Masukkan nilai b dalam float

# hitung hasil penambahan disini
# hitung hasil pengurangan disini
# hitung hasil perkalian disini
# hitung hasil pembagian disini


print("\nYeahhhh")
```

````python
# Masukkan nilai a dalam float
a = float(input("Masukkan nilai a: "))
# Masukkan nilai b dalam float
b = float(input("Masukkan nilai b: "))

# hitung hasil penambahan
penjumlahan = a + b
# hitung hasil pengurangan
pengurangan = a - b
# hitung hasil perkalian
perkalian = a * b
# hitung hasil pembagian
if b != 0:
    pembagian = a / b
else:
    pembagian = "Tak terdefinisi (b=0)"

# tampilkan hasil
print("\nHasil Penjumlahan:", penjumlahan)
print("Hasil Pengurangan:", pengurangan)
print("Hasil Perkalian:", perkalian)
print("Hasil Pembagian:", pembagian)

print("\nTerimakasih!")
````
