```python
# Import library numpy, pandas dan scikit-learn
import numpy as np
import pandas as pd
from sklearn import tree

print("Import berhasil")
```

```python
# Membaca Dataset dari File ke Pandas DataFrame
irisDataset = pd.read_csv('Dataset Iris.csv', delimiter=';', header=0)

print(irisDataset.head())
```

```python
print(irisDataset.info())
```

```python
# Cek data secara ringkas
print(irisDataset.head())
print("\nKolom dataset:", irisDataset.columns)
```

```python
# Mengubah kelas (kolom "Species") dari String ke Unique-Integer
irisDataset["Species"] = pd.factorize(irisDataset["Species"])[0]

print(irisDataset.head())
```

```python
irisDataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Hapus kolom Id kalau ada (biar gak error)
irisDataset = irisDataset.drop(columns=[col for col in ["Id"] if col in irisDataset.columns])

print(irisDataset.head())
```

```python
print(irisDataset.columns)
```

```python
irisDataset
```

```python
# Mengubah DataFrame ke array NumPy
irisDataset = irisDataset.to_numpy()

print(type(irisDataset))
print(irisDataset[:5])
```

```python
# Membagi Dataset => 80 training, 20 testing (manual split)
dataTraining = np.concatenate((irisDataset[0:40, :], irisDataset[50:90, :]), axis=0)
dataTesting = np.concatenate((irisDataset[40:50, :], irisDataset[90:100, :]), axis=0)

print("Data training shape:", dataTraining.shape)
print("Data testing shape:", dataTesting.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Memecah Dataset ke Input dan Label
inputTraining = dataTraining[:, 0:4]
inputTesting = dataTesting[:, 0:4]
labelTraining = dataTraining[:, 4]
labelTesting = dataTesting[:, 4]

print("Shape inputTraining:", inputTraining.shape)
print("Shape inputTesting:", inputTesting.shape)
print("Shape labelTraining:", labelTraining.shape)
print("Shape labelTesting:", labelTesting.shape)
```

```python
print(inputTesting)
```

```python
print(inputTesting.shape)
```

```python
# Mendefinisikan Decision Tree Classifier
model = tree.DecisionTreeClassifier()

print(model)
```

```python
# Mentraining Model
model = model.fit(inputTraining, labelTraining)

print("Model berhasil dilatih")
```

```python
hasilPrediksi = model.predict(inputTesting)
print(hasilPrediksi)
```

```python
# Perbandingan Hasil
print("\nPerbandingan (Asli vs Prediksi):")
for i in range(len(labelTesting)):
    print(f"Data ke-{i+1}: Asli={labelTesting[i]}, Prediksi={hasilPrediksi[i]}")
```

```python
# Memprediksi Input Data Testing
hasilPrediksi = model.predict(inputTesting)

print("Label Sebenarnya :", labelTesting)
print("Hasil Prediksi   :", hasilPrediksi)

# Tambahan cek sederhana
print("Jumlah data test :", len(labelTesting))
```

```python
from sklearn.metrics import accuracy_score
print("Akurasi:", accuracy_score(labelTesting, hasilPrediksi))
```

```python
# Menghitung Akurasi
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()

total = prediksiBenar + prediksiSalah

print("Prediksi Benar :", prediksiBenar, "data")
print("Prediksi Salah :", prediksiSalah, "data")
print("Akurasi :", round((prediksiBenar / total) * 100, 2), "%")
```

```python
# Kesimpulan Akurasi
akurasi = prediksiBenar / (prediksiBenar + prediksiSalah) * 100

print("\nKesimpulan:")
print(f"Model memiliki tingkat akurasi sebesar {round(akurasi, 2)}%")
print("Model cukup baik untuk klasifikasi dataset Iris.")
```
